Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Ans-The Type I and Type II Errors in Hypothesis Testing are given below :-

When we perform a hypothesis test, we start with two statements:

Null hypothesis (H₀): There is no effect or difference (status quo).

Alternative hypothesis (H₁ or Ha): There is an effect or difference.

When we make a decision based on sample data, we can either reject or fail to reject the null hypothesis.
However, since our decision is based on probability, two types of errors can occur.

1. Type I Error (False Positive)

Definition: Occurs when we reject the null hypothesis even though it is true.

Symbol: α (alpha)

Also known as: The significance level of the test.

Example:
Suppose a medical test is used to detect a disease.

H₀: The patient does not have the disease.

H₁: The patient has the disease.

If the test shows the patient has the disease when they actually don’t, that’s a Type I error.

Impact:
You conclude there is an effect or difference when in reality, there isn’t.
This can lead to false alarms — for example, approving an ineffective drug.

2. Type II Error (False Negative)

Definition: Occurs when we fail to reject the null hypothesis even though it is false.

Symbol: β (beta)

Example:
Using the same medical example:
If the test shows the patient does not have the disease when they actually do, that’s a Type II error.

Impact:
You miss a real effect or difference — a missed detection — for example, failing to recognize that a new treatment actually works.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Ans-
The P-value (probability value) is a key concept used to decide whether to reject the null hypothesis (H₀) in a hypothesis test.

*__Definition__*

The P-value is the probability of obtaining a test statistic (or result) as extreme as, or more extreme than, the one observed in your sample, assuming that the null hypothesis is true.

*__Interpretation__*

A small P-value means that the observed data are unlikely under the null hypothesis.

A large P-value means that the observed data are consistent with the null hypothesis.

*__Important points__*

- The P-value does not prove the null or alternative hypothesis — it only indicates how consistent the data is with H₀.

- A low P-value (< α) → evidence against H₀.

- A high P-value (> α) → evidence for H₀ (or insufficient evidence against it).

- Common significance levels: α = 0.05, 0.01, or 0.10.

Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

Ans-

Z-Test vs. T-Test in Hypothesis Testing

Both Z-tests and T-tests are statistical methods used to compare sample data with population parameters or between two samples — but they differ based on sample size and what is known about the population.

1. Z-Test

Definition:
A Z-test is used when the population variance (σ²) or standard (\sigma) deviation (σ) is known, and the sample size is large (typically n ≥ 30).

When to Use a Z-Test:

Population standard deviation is known.

Sample size is large (n ≥ 30).

Data follows a normal distribution (or approximately normal due to Central Limit Theorem).

Examples:

Testing if the average height of students differs from a known population mean when σ is known.

Quality control testing in manufacturing where population parameters are known.

2. T-Test

Definition:
A T-test is used when the population variance (σ²) or standard deviation (σ) is unknown, and the sample size is small (n < 30).

When to Use a T-Test:

Population standard deviation is unknown.

Sample size is small (n < 30).

Data is approximately normally distributed.

Types of T-Tests:

One-sample T-test – compares a sample mean to a known population mean.

Independent (two-sample) T-test – compares means of two independent groups.

Paired T-test – compares means from the same group at two different times (before/after).

Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

Ans- Confidence Interval (CI)

A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.

- Definition

A confidence interval gives an estimated range of values which is likely to include the true population value, based on a sample statistic.

It is expressed as:

Confidence Interval = Sample Statistic  ± Margin of Error

Factors affecting the margin of error:

Factor	               Effect on Margin of Error

. Sample size (n)	    Larger sample → smaller margin → narrower CI

. Confidence level (e.g., 95% vs 99%)	Higher confidence → larger margin → wider CI

. Variability (σ or s)	More variability → larger margin → wider CI

- Interpretation

A narrow confidence interval → more precise estimate (less uncertainty).

A wide confidence interval → less precise estimate (more uncertainty).

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Ans- ANOVA, short for Analysis of Variance, is a statistical test used to compare the means of three or more groups to see if at least one of them is significantly different from the others.

*__Purpose of ANOVA__*

The purpose of ANOVA is to determine whether differences among sample means are due to actual differences between groups or merely due to random chance.

*__ANOVA Extends Hypothesis Testing__*

If ANOVA finds a significant F-value, it tells us that not all means are equal, but it doesn’t tell which ones differ — for that, we use post-hoc tests like:

- Tukey’s HSD

- Bonferroni correction

- Scheffé test

```python
# Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset

import numpy as np
from scipy import stats

# --- Sample Data ---
# Suppose we have a sample of exam scores
data = [50, 55, 60, 54, 63, 66, 65, 70, 86, 90]

# Population mean (μ0)
population_mean = 65

# Population standard deviation (σ)
population_std = 4

# Sample statistics
sample_mean = np.mean(data)
sample_size = len(data)

# --- Z-test formula ---
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# --- Display results ---
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

# --- Interpretation ---
alpha = 0.05
if p_value < alpha:
    print(" Reject the null hypothesis (H0): There is a significant difference.")
else:
    print(" Fail to reject the null hypothesis (H0): No significant difference.")

```

```python
# Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for a sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default = 0.95)
    """
    # --- Step 1: Compute basic statistics ---
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)   # sample standard deviation
    n = len(data)

    # --- Step 2: Compute standard error ---
    standard_error = sample_std / np.sqrt(n)

    # --- Step 3: Find t-critical value for given confidence ---
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # --- Step 4: Compute margin of error and CI ---
    margin_of_error = t_critical * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # --- Step 5: Print results ---
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Level: {confidence*100:.1f}%")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    # --- Step 6: Visualization ---
    plt.figure(figsize=(12,7))
    plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', capsize=8, color='blue')
    plt.axhline(y=sample_mean, color='gray', linestyle='--', label='Sample Mean')
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Value")
    plt.xticks([])
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# --- Example usage ---
# Generate random sample data (e.g., exam scores)
np.random.seed(50)
data = np.random.normal(loc=75, scale=10, size=50)

plot_confidence_interval(data, confidence=0.95)
```

```python
# Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

# step 1 - define observed and expected distribution
# example = A dice rolled out 40 times and we count how many times each face appear
observed = np.array([16, 18, 16, 14, 12, 12])
total_observed = np.sum(observed)
# The sum of expected frequencies must equal the sum of observed frequencies
# For a fair six-sided die rolled 40 times, the expected frequency for each face is 40/6
expected = np.full(len(observed), total_observed / len(observed))


# step 2 - perform chi-square goodness-of-fit test
chi2_statistic, p_value = chisquare(f_obs=observed, f_exp=expected)

# step 3 - display results
print(f"Chi-square Statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# step 4 - interpritation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed and expected distributions are not the same.")
else:
    print("Fail to reject the null hypothesis: Observed and expected distributions are the same.")
```
